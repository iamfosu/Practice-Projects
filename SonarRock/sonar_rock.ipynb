```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
#svm import
from sklearn.svm import SVC
```

```python
X = sonardata.drop(columns=["R"], axis=1)
Y = sonardata['R']
X
```

```python
Y
```

Training and Testing dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=10)
#X_train, Xtest, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=10)
```

```python
X_train.shape, Y_train.shape, Y_test.shape
```

Model

```python
# Logistic Regression Model
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, Y_train)
```

Logistic Regression 
Model Evaluation

```python
# Logistic Regression accuracy on training set
X_train_predicted = lr_model.predict(X_train)
training_accuracy = accuracy_score(X_train_predicted, Y_train)
training_accuracy
```

```python
#accuracy in percentage
training_accuracy_percentage = training_accuracy * 100 
print(f"Training Accuracy : {training_accuracy_percentage}%")
```

```python
#accuracy on test set
X_test_predicted = lr_model.predict(X_test)
test_accuracy = accuracy_score(X_test_predicted, Y_test)
test_accuracy
```

```python
#Test accuracy percentage
test_accuracy_percentage = test_accuracy * 100
print(f"Test Accuracy : {test_accuracy_percentage}%")
```

SVM model evaluation

```python
# SVM model
svm_model = SVC(kernel='linear')   # kernal='linear' is used for linearly separable data 
```

```python
svm_model.fit(X_train, Y_train)
```

```python
# training accuracy for SVM 
training_accuracy_svm = svm_model.fit(X_train, Y_train)
```

```python
training_accuracy_svm_score = svm_model.score(X_train, Y_train)
print(f"Training accuracy for SVM : {training_accuracy_svm_score * 100}%")
```

```python
# test accuracy for SVM
test_accuracy_svm_score = svm_model.score(X_test, Y_test)
print(f"Test accuracy score for SVM: {test_accuracy_svm_score * 100}%")
```
